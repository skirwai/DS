```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

# This just a scaler techinuq to have less varience for x axis

```python
scaler.fit(df.drop(['TARGET CLASS'],axis=1))
```

```python
df_scaler = scaler.transform(df.drop(['TARGET CLASS'],axis=1))
```

```python
df_scaler
```

```python
df_transformed = pd.DataFrame(data=df_scaler,columns=df.columns[:-1])
```

```python
df_transformed.mean()
```

# Above means is for reduce the varience

```python
df_transformed.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_transformed, df['TARGET CLASS'], test_size=0.4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,pred))
```

# Now choosing a K Value from iteration 

```python
error_rate = []

for i in range(1,50):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pre_i = knn.predict(X_test)
    error_rate.append(np.mean(pre_i != y_test))
```

```python

plt.figure(figsize=(10,4))
plt.plot(range(1,50),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
